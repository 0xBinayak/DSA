```python
class Node:
    
    def __init__(self,value):
        self.data=value
        self.next=None
```

```python
class Stack :
    
    def __init__(self):
        self.top = None
    
    
    def  isempty(self):
        
        return  self.top == None
    
    
    def push(self,value):
        
        new_node= Node(value)
        
        new_node.next = self.top
        
        self.top=  new_node
        
    
    
    def traverse(self):
        
        temp= self.top
        
        while self.top != None :
            
            
            print (temp.data)
            
            temp=temp.next
            
            
    def peak(self):
        
        if (self.isempty()):
            return "Stack empty"     
        
        else:
            
            return self.top.data
        
        
        
    def pop(self):
        
        
        if (self.isempty()):
            return "Stack empty"
        
        else:
            data= self.top
            self.top=self.top.next 
            return data   
        
```
